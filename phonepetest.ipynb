```python
#aggre_insurance
path7= "D:/INTERNSHIP LABMENTIX/Phonepe 2 (22 JUNE TO 10 JULY 2025)/pulse/data/aggregated/insurance/country/india/state/"

agg_insur_list= os.listdir(path7)

columns7= {"States":[], "Years":[], "Quarter":[], "Insurance_type":[], "Insurance_count":[],"Insurance_amount":[] }

for state in agg_insur_list:
    cur_states =path7+state+"/"
    agg_year_list = os.listdir(cur_states)
    
    for year in agg_year_list:
        cur_years = cur_states+year+"/"
        agg_file_list = os.listdir(cur_years)

        for file in agg_file_list:
            cur_files = cur_years+file
            data = open(cur_files,"r")
            A = json.load(data)

            for i in A["data"]["transactionData"]:
                name = i["name"]
                count = i["paymentInstruments"][0]["count"]
                amount = i["paymentInstruments"][0]["amount"]
                columns7["Insurance_type"].append(name)
                columns7["Insurance_count"].append(count)
                columns7["Insurance_amount"].append(amount)
                columns7["States"].append(state)
                columns7["Years"].append(year)
                columns7["Quarter"].append(int(file.strip(".json")))


aggre_insurance = pd.DataFrame(columns7)

aggre_insurance["States"] = aggre_insurance["States"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
aggre_insurance["States"] = aggre_insurance["States"].str.replace("-"," ")
aggre_insurance["States"] = aggre_insurance["States"].str.title()
aggre_insurance['States'] = aggre_insurance['States'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")
```

```python

# Step 1: Connect to MySQL database
connection = pymysql.connect(
    host='localhost',
    user='root',       # 🔁 Replace with your MySQL username
    password='123456',   # 🔁 Replace with your MySQL password
    database='phonepe_data',
    port=3306
)

# Step 2: List of table names to fetch
table_names = [
    'aggregated_insurance',
    'aggregated_transaction',
    'aggregated_user',
    'map_insurance',
    'map_transaction',
    'map_user',
    'top_insurance',
    'top_transaction',
    'top_user'
]

# Step 3: Dictionary to hold DataFrames
dataframes = {}

# Step 4: Fetch all tables into DataFrames
for table in table_names:
    query = f"SELECT * FROM {table}"
    df = pd.read_sql(query, connection)
    dataframes[table] = df
    print(f"✅ Loaded table: {table} → DataFrame shape: {df.shape}")

# Step 5: Optional: Assign individual names to each DataFrame
aggregated_insurance_df = dataframes['aggregated_insurance']
aggregated_transaction_df = dataframes['aggregated_transaction']
aggregated_user_df = dataframes['aggregated_user']
map_insurance_df = dataframes['map_insurance']
map_transaction_df = dataframes['map_transaction']
map_user_df = dataframes['map_user']
top_insurance_df = dataframes['top_insurance']
top_transaction_df = dataframes['top_transaction']
top_user_df = dataframes['top_user']

# Step 6: Create folder for CSV exports
output_path = output_path = r'D:\INTERNSHIP LABMENTIX\Phonepe 2 (22 JUNE TO 10 JULY 2025)\ACTUAL PROJECT\DATASETS'

os.makedirs(output_path, exist_ok=True)

# Step 7: Export each DataFrame to CSV
for table_name, df in dataframes.items():
    file_path = os.path.join(output_path, f"{table_name}.csv")
    df.to_csv(file_path, index=False)
    print(f"📁 Exported: {file_path}")

# Step 8: Close connection
connection.close()
print("✅ All tables exported successfully to CSV.")
```
